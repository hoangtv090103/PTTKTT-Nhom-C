```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu từ bảng
problems = [f'Problem {i}' for i in range(1, 11)]
greedy_means = [5907, 5704, 173, 41, 3570, 8344, 2865, 1029, 467, 8184]
dp_means = [112, 155, 368, 995, 1585, 2151, 2221, 2588, 2836, 3239]
ga_means = [20336, 24578, 49400, 13319, 36115, 39851, 38196, 16017, 30598, 40202]

# Thiết lập vị trí cho các thanh
x = np.arange(len(problems))

# Chiều rộng của các thanh
width = 0.25

# Tạo biểu đồ
fig, ax = plt.subplots(figsize=(12, 6))

# Vẽ các thanh cho từng thuật toán
bars1 = ax.bar(x - width, greedy_means, width, label='Tham lam (Greedy)', color='lightblue')
bars2 = ax.bar(x, dp_means, width, label='Quy hoạch động (Dynamic Programming)', color='lightgreen')
bars3 = ax.bar(x + width, ga_means, width, label='Di truyền (Genetic Algorithm)', color='salmon')

# Thêm nhãn và tiêu đề
ax.set_xlabel('Bài toán')
ax.set_ylabel('Thời gian trung bình (ms)')
ax.set_title('So sánh thời gian trung bình của các thuật toán cho từng bài toán')
ax.set_xticks(x)
ax.set_xticklabels(problems)
ax.legend()

# Hiển thị giá trị trên các thanh
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 điểm ở trên thanh
                    textcoords="offset points",
                    ha='center', va='bottom')

add_value_labels(bars1)
add_value_labels(bars2)
add_value_labels(bars3)

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu từ bảng
problems = [f'Problem {i}' for i in range(1, 11)]
greedy_means = [50, 52, 18, 33.3, 119.4, 112, 104.8, 52.9, 103, 133]
dp_means = [50, 50, 20, 24, 92, 72, 57, 23, 130, 110]
ga_means = [90, 75.5, 39, 34, 234, 120.4, 115.6, 87.6, 174.3, 143]
manual_values = [100, 90, 40, 34, 240, 130, 118, 100, 190, 210]  # Giá trị mong muốn

# Thiết lập vị trí cho các thanh
x = np.arange(len(problems))

# Chiều rộng của các thanh
width = 0.2

# Tạo biểu đồ
fig, ax = plt.subplots(figsize=(12, 6))

# Vẽ các thanh cho từng thuật toán
bars1 = ax.bar(x - width, greedy_means, width, label='Tham lam (Greedy)', color='lightblue')
bars2 = ax.bar(x, dp_means, width, label='Quy hoạch động (Dynamic Programming)', color='lightgreen')
bars3 = ax.bar(x + width, ga_means, width, label='Di truyền (Genetic Algorithm)', color='salmon')
bars4 = ax.bar(x + 1.5 * width, manual_values, width, label='Giá trị mong muốn (Manual)', color='orange')  # Cột cho giá trị mong muốn

# Thêm nhãn và tiêu đề
ax.set_xlabel('Bài toán')
ax.set_ylabel('Giá trị trung bình')
ax.set_title('So sánh giá trị trung bình của các thuật toán cho từng bài toán')
ax.set_xticks(x + 0.5 * width)  # Điều chỉnh vị trí nhãn
ax.set_xticklabels(problems)
ax.legend()

# Hiển thị giá trị trên các thanh
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 điểm ở trên thanh
                    textcoords="offset points",
                    ha='center', va='bottom')

add_value_labels(bars1)
add_value_labels(bars2)
add_value_labels(bars3)
add_value_labels(bars4)  # Thêm giá trị cho cột manual

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```
