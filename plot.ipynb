```python
import matplotlib.pyplot as plt
import numpy as np

# D liu t bng
problems = [f'Problem {i}' for i in range(1, 11)]
greedy_means = [5907, 5704, 173, 41, 3570, 8344, 2865, 1029, 467, 8184]
dp_means = [112, 155, 368, 995, 1585, 2151, 2221, 2588, 2836, 3239]
ga_means = [20336, 24578, 49400, 13319, 36115, 39851, 38196, 16017, 30598, 40202]

# Thit lp v tr cho cc thanh
x = np.arange(len(problems))

# Chiu rng ca cc thanh
width = 0.25

# To biu 
fig, ax = plt.subplots(figsize=(12, 6))

# V cc thanh cho tng thut ton
bars1 = ax.bar(x - width, greedy_means, width, label='Tham lam (Greedy)', color='lightblue')
bars2 = ax.bar(x, dp_means, width, label='Quy hoch ng (Dynamic Programming)', color='lightgreen')
bars3 = ax.bar(x + width, ga_means, width, label='Di truyn (Genetic Algorithm)', color='salmon')

# Thm nhn v tiu 
ax.set_xlabel('Bi ton')
ax.set_ylabel('Thi gian trung bnh (ms)')
ax.set_title('So snh thi gian trung bnh ca cc thut ton cho tng bi ton')
ax.set_xticks(x)
ax.set_xticklabels(problems)
ax.legend()

# Hin th gi tr trn cc thanh
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 im  trn thanh
                    textcoords="offset points",
                    ha='center', va='bottom')

add_value_labels(bars1)
add_value_labels(bars2)
add_value_labels(bars3)

# Hin th biu 
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# D liu t bng
problems = [f'Problem {i}' for i in range(1, 11)]
greedy_means = [50, 52, 18, 33.3, 119.4, 112, 104.8, 52.9, 103, 133]
dp_means = [50, 50, 20, 24, 92, 72, 57, 23, 130, 110]
ga_means = [90, 75.5, 39, 34, 234, 120.4, 115.6, 87.6, 174.3, 143]
manual_values = [100, 90, 40, 34, 240, 130, 118, 100, 190, 210]  # Gi tr mong mun

# Thit lp v tr cho cc thanh
x = np.arange(len(problems))

# Chiu rng ca cc thanh
width = 0.2

# To biu 
fig, ax = plt.subplots(figsize=(12, 6))

# V cc thanh cho tng thut ton
bars1 = ax.bar(x - width, greedy_means, width, label='Tham lam (Greedy)', color='lightblue')
bars2 = ax.bar(x, dp_means, width, label='Quy hoch ng (Dynamic Programming)', color='lightgreen')
bars3 = ax.bar(x + width, ga_means, width, label='Di truyn (Genetic Algorithm)', color='salmon')
bars4 = ax.bar(x + 1.5 * width, manual_values, width, label='Gi tr mong mun (Manual)', color='orange')  # Ct cho gi tr mong mun

# Thm nhn v tiu 
ax.set_xlabel('Bi ton')
ax.set_ylabel('Gi tr trung bnh')
ax.set_title('So snh gi tr trung bnh ca cc thut ton cho tng bi ton')
ax.set_xticks(x + 0.5 * width)  # iu chnh v tr nhn
ax.set_xticklabels(problems)
ax.legend()

# Hin th gi tr trn cc thanh
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 im  trn thanh
                    textcoords="offset points",
                    ha='center', va='bottom')

add_value_labels(bars1)
add_value_labels(bars2)
add_value_labels(bars3)
add_value_labels(bars4)  # Thm gi tr cho ct manual

# Hin th biu 
plt.tight_layout()
plt.show()
```
